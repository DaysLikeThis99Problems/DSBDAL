```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
```

```python
print("Features and their types:")
print("Features: ", df.columns[:-1].tolist())  # Exclude 'species' column as it's categorical
print("Feature types:")
for feature in df.columns[:-1]:
    print(f"{feature}: {df[feature].dtype}")
```

```python
df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms for Each Feature in Iris Dataset')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, feature in enumerate(df.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[feature], kde=True, color='lightblue', bins=20)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
for feature in df.columns[:-1]:
    feature_data = df[feature]
    skewness = feature_data.skew()
    min_val = feature_data.min()
    max_val = feature_data.max()

    print(f"Feature: {feature}")
    print(f"Skewness: {skewness:.2f}")
    print(f"Min Value: {min_val}")
    print(f"Max Value: {max_val}")
    print(f"Outliers: Min value = {min_val}, Max value = {max_val}")
    print("-" * 50)

    Q1 = feature_data.quantile(0.25)
    Q3 = feature_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = feature_data[(feature_data < lower_bound) | (feature_data > upper_bound)]

    print(f"Outliers: {outliers}")
    print("=" * 50)
```
