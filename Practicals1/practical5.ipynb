```python
# 1. Import Libraries
import pandas as pd
import numpy as np
```

```python
# Load the iris dataset
# df = pd.read_csv(r"D:\DSBDA-PS-Solutions-main\Datasets\Iris.xls")
# df.head()
#or
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
```

```python
grouped_stats = df.groupby('species').agg({
    'sepal_length': ['mean', 'median', 'min', 'max', 'std'],
    'sepal_width': ['mean', 'median', 'min', 'max', 'std'],
    'petal_length': ['mean', 'median', 'min', 'max', 'std'],
    'petal_width': ['mean', 'median', 'min', 'max', 'std'],
})

grouped_stats
```

```python
# Convert 'species' to numeric values
species_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df['species_num'] = df['species'].map(species_map)

# List containing numeric values for each response
species_numeric_list = df['species_num'].tolist()
print("Numeric list for species:", species_numeric_list[:10])  # print first 10 as sample
```

```python
# Filtering each species
setosa_df = df[df['species'] == 'setosa']
versicolor_df = df[df['species'] == 'versicolor']
virginica_df = df[df['species'] == 'virginica']

# Display basic statistics
print("Setosa Statistics:\n", setosa_df.describe())
print("\nVersicolor Statistics:\n", versicolor_df.describe())
print("\nVirginica Statistics:\n", virginica_df.describe())
```
