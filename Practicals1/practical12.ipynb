```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display first few rows
print(iris.head())
```

```python
print("\n--- List of Features ---")
print(iris.columns)
print("\n--- Features and Their Types ---")
print(iris.dtypes)
```

```python
iris.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle('Histograms for Each Feature in Iris Dataset')
plt.show()
```

```python
features = iris.columns[:-1]
print(features)
```

```python
plt.figure(figsize=(10, 6))
for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(y=iris[feature], color='lightblue')
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    plt.hist(iris[feature], bins=15, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
