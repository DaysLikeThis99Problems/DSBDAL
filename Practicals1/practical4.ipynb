```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\sumit\OneDrive\Desktop\Student_performance.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Example: fill missing values in 'math score' with mean
df['math_score'] = df['math_score'].fillna(df['math_score'].mean())

# Example: drop rows with missing values in 'reading score'
df.dropna(subset=['reading_score'], inplace=True)
```

```python
df['math_score'].skew()
```

```python
df['reading_score'].skew()
```

```python
df['writing_score'].skew()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math_score'], kde=True)
plt.title('Distribution of Math Score')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['reading_score'], kde=True)
plt.title('Distribution of reading Score')
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(df['writing_score'], kde=True)
plt.title('Distribution of writing Score')
plt.show()
```

```python
df['math_score power'] = df['math_score']**1.5
print("Math score skewness after applying Power transformation: ", df['math_score power'].skew())

df['reading_score power'] = df['reading_score']**1.5
print("Reading score skewness after applying Power transformation: ",df['reading_score power'].skew())

df['writing_score power'] = df['writing_score']**1.5
print("Writing score skewness after applying Power transformation: ",df['writing_score power'].skew())
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['math_score power'],kde=True)
plt.title("Math after removing skewness")
plt.show()
```
