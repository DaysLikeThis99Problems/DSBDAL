```python
import pandas as pd
df = pd.read_csv(r"C:\Users\sumit\OneDrive\Desktop\sample_weather.txt")
df.head()
```

```python
print(df.columns.tolist())
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['DATE']=pd.to_datetime(df['DATE'])
```

```python
df['Year'] = df['DATE'].dt.year
df['Month'] = df['DATE'].dt.month
```

```python
year_avg = df.groupby('Year')[['TEMP','DEWP','WDSP']].mean()
month_avg = df.groupby('Month')[['TEMP','DEWP','WDSP']].mean()
```

```python
print("Year-wise Averages:\n", year_avg)
print("\nMonth-wise Averages:\n", month_avg)
```
